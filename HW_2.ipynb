```
import pandas as pd
raw_url = 'https://raw.githubusercontent.com/shaksiper/db/main/cities.csv'
citiesDataFrame = pd.read_csv(raw_url, skipinitialspace = True, quotechar = '"')
```

```
citiesDataFrame.head(10)
```

```
citiesDataFrame.tail(10)
```

Below is an example for instertion a row/tuple into the table

```
#Insert row/tuple

new_row = {'LatD' : 135,
           'LatM' : 50,
           'LatS' : 10,
           'NS'   : 'S',
           'LonD' : 100,
           'LonM' : 100,
           'LonS' : 30,
           'EW'   : 'E',
           'City' : 'Aydin',
           'State': 'TR'}
citiesDataFrame = citiesDataFrame.append(new_row, ignore_index=True)
```

```
citiesDataFrame.tail(10)
```

Below is an example for updating a row/tuple

```
citiesDataFrame.loc[citiesDataFrame['City'] == 'Reading', ['NS', 'EW']] = ['S', 'E']
citiesDataFrame.tail(10)
```

Below is an exaple of delete/drop of a specific row/tuple

```
citiesDataFrame.shape
```

```
where = ((citiesDataFrame.LatD == 50) & (citiesDataFrame.LatM == 25) & (citiesDataFrame.LatS == 11))
#124 	50 	25 	11 	N 	104 	39 	0 	W 	Regina 	SA
#The row is deleted
citiesDataFrame = citiesDataFrame[~where]
citiesDataFrame.shape
```
